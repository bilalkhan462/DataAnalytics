### Deleting suffix from duration and change its data type from object to integer

```python
ride['duration']
```

```python
# Deleting suffix "minutes"
ride['duration_trim'] = ride['duration'].str.strip('minutes')

# Convert data type
ride['duration_time'] = ride['duration_trim'].astype('int')

# Assert statement to confirm datatype
assert ride['duration_time'].dtype == 'int'

# Print all alternations
print(ride[['duration','duration_trim','duration_time']])
print(ride[['duration','duration_trim','duration_time']].dtypes)
print(ride['duration_time'].mean())
```

### Delete additional columns

```python
ride.drop(['duration','duration_trim'], axis=1, inplace = True)
ride.head(3)
```
